```python
print("Ramya R")
print("24BAD096")
# Import Libraries
import numpy as np
import matplotlib.pyplot as plt

# Generate Input Values
x = np.linspace(-10, 10, 500)

# Activation Functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
def tanh(x):
    return np.tanh(x)
def relu(x):
    return np.maximum(0, x)

# Compute Outputs
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)
y_relu = relu(x)

# Plot Activation Functions
plt.figure(figsize=(12,6))

plt.plot(x, y_sigmoid, label='Sigmoid', linewidth=2)
plt.plot(x, y_tanh, label='Tanh', linewidth=2)
plt.plot(x, y_relu, label='ReLU', linewidth=2)

plt.title("Activation Functions")
plt.xlabel("Input (x)")
plt.ylabel("Output")
plt.grid(True)
plt.legend()
plt.show()

# Comparison Table
print("\nComparison of Activation Functions\n")

print("{:<12} {:<20} {:<25} {:<25} {:<25}".format(
    "Function",
    "Output Range",
    "Saturation Region",
    "Gradient Behavior",
    "Typical Applications"
))

print("-"*120)

print("{:<12} {:<20} {:<25} {:<25} {:<25}".format(
    "Sigmoid",
    "(0,1)",
    "Large +ve and -ve inputs",
    "Vanishing Gradient",
    "Binary Classification Output"
))

print("{:<12} {:<20} {:<25} {:<25} {:<25}".format(
    "Tanh",
    "(-1,1)",
    "Large +ve and -ve inputs",
    "Vanishing Gradient",
    "Hidden Layers"
))

print("{:<12} {:<20} {:<25} {:<25} {:<25}".format(
    "ReLU",
    "[0,∞)",
    "Negative inputs only",
    "No Vanishing Gradient for +ve values",
    "Deep Neural Networks"
))

# Computational Efficiency
print("\nComputational Efficiency")
print("-------------------------")
print("Sigmoid : Slow (uses exponential function)")
print("Tanh    : Moderate (uses exponential function)")
print("ReLU    : Fast (simple max operation)")
```

```python
print("Ramya R")
print("24BAD096")
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt
import pandas as pd

# Load MNIST Dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalize Images
x_train = x_train / 255.0
x_test = x_test / 255.0

# Function to Create Model
def create_model(activation):
    model = Sequential([
        Flatten(input_shape=(28,28)),
        Dense(128, activation=activation),
        Dense(64, activation=activation),
        Dense(10, activation='softmax')
    ])

    model.compile(
        optimizer='adam',
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )

    return model

# Activation Functions to Compare
activations = ['sigmoid', 'tanh', 'relu']

results = {}

# Train Models
for act in activations:

    print("\nTraining Model with", act.upper(), "Activation")

    model = create_model(act)

    history = model.fit(
        x_train,
        y_train,
        epochs=5,
        batch_size=128,
        validation_split=0.2,
        verbose=1
    )

    results[act] = history

# Plot Accuracy
plt.figure(figsize=(10,5))

for act in activations:
    plt.plot(results[act].history['accuracy'], label=act)

plt.title("Training Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.show()

# Plot Validation Accuracy
plt.figure(figsize=(10,5))

for act in activations:
    plt.plot(results[act].history['val_accuracy'], label=act)

plt.title("Validation Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.show()

# Plot Training Loss
plt.figure(figsize=(10,5))

for act in activations:
    plt.plot(results[act].history['loss'], label=act)

plt.title("Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.show()

# Plot Validation Loss
plt.figure(figsize=(10,5))

for act in activations:
    plt.plot(results[act].history['val_loss'], label=act)

plt.title("Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.show()

# Performance Comparison Table

comparison = []

for act in activations:

    h = results[act].history

    comparison.append([
        act.upper(),
        round(h['accuracy'][-1],4),
        round(h['val_accuracy'][-1],4),
        round(h['loss'][-1],4),
        round(h['val_loss'][-1],4),
        len(h['accuracy'])
    ])

df = pd.DataFrame(comparison,
                  columns=[
                      "Activation Function",
                      "Training Accuracy",
                      "Validation Accuracy",
                      "Training Loss",
                      "Validation Loss",
                      "Epochs"
                  ])

print("\nPerformance Comparison\n")
print(df)
```

```python
print("Ramya R- 24BAD096")
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt
import pandas as pd

# Load MNIST Dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

# Normalize Data
x_train = x_train.reshape(-1, 784).astype("float32") / 255.0
x_test = x_test.reshape(-1, 784).astype("float32") / 255.0

# Function to Build and Train Model
def build_and_train(optimizer, name):

    model = keras.Sequential([
        layers.Input(shape=(784,)),
        layers.Dense(128, activation='relu'),
        layers.Dense(64, activation='relu'),
        layers.Dense(10, activation='softmax')
    ])

    model.compile(
        optimizer=optimizer,
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )

    print("Training using", name)

    history = model.fit(
        x_train,
        y_train,
        validation_data=(x_test, y_test),
        epochs=5,
        batch_size=128,
        verbose=0
    )

    return history

# Optimizers
optimizers = {
    "SGD": keras.optimizers.SGD(),
    "Momentum": keras.optimizers.SGD(momentum=0.9),
    "RMSProp": keras.optimizers.RMSprop(),
    "Adam": keras.optimizers.Adam()
}

histories = {}

# Train Models
for name, opt in optimizers.items():
    histories[name] = build_and_train(opt, name)

# Accuracy Graph

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)

for name in optimizers.keys():
    plt.plot(histories[name].history['accuracy'], label=name)

plt.title("Training Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)

plt.subplot(1,2,2)

for name in optimizers.keys():
    plt.plot(histories[name].history['val_accuracy'], label=name)

plt.title("Validation Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Loss Graph

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)

for name in optimizers.keys():
    plt.plot(histories[name].history['loss'], label=name)

plt.title("Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)

plt.subplot(1,2,2)

for name in optimizers.keys():
    plt.plot(histories[name].history['val_loss'], label=name)

plt.title("Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Comparison Table

results = []

for name in optimizers.keys():

    h = histories[name].history

    results.append([
        name,
        round(h['loss'][-1],4),
        round(h['val_loss'][-1],4),
        len(h['loss']),
        round(h['accuracy'][-1],4),
        round(h['val_accuracy'][-1],4)
    ])

df = pd.DataFrame(results, columns=[
    "Optimizer",
    "Training Loss",
    "Validation Loss",
    "Epochs",
    "Training Accuracy",
    "Validation Accuracy"
])

print("\nPerformance Comparison\n")
print(df)
```
